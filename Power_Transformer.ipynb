# Load Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('concrete_data.csv')
df.sample(3)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#Input out split
x=df.drop('Strength',axis=1)
y=df['Strength']
```

```python
#Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Linear Regression without Transform and scaling


```python
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
accuracy = r2_score(y_test,y_pred)
accuracy
```

# Checking Cross validation

```python
lr = LinearRegression()

np.mean(cross_val_score(lr,x,y,scoring='r2'))
```

#  PDF and QQ PLOT for all Columns

```python
from statsmodels.graphics.gofplots import ProbPlot # import for QQ plot
import scipy.stats as stats

for col in x_train.columns:
    print(col," PDF vs QQPlot ")
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(1,2,2)
    ProbPlot(x_train[col], dist=stats.norm).qqplot(line='s', ax=plt.gca())
    plt.title(col)
    # ProbPlot(x_train[col],dist='norm',plot=plt)
    plt.show()
```

## Power Transformer by default use Yeo-Johnson and by default use Standardization scaling
parameters:


1.   method : {'yeo-johnson','box-cox'}, default='yeo-johnson'
2.   tandardize : bool , default=True
3.   copy: bool, default=True  --(set false to inplace the computation during transformation

# Applying Box-Cox

```python

# Applying Box-Cox
pt = PowerTransformer(method='box-cox',standardize=True)
x_train_pt = pt.fit_transform(x_train+0.000001)   # adding some value as box-cox works for only positive integers but we have zero in the data
x_test_pt = pt.transform(x_test+0.000001)


pd.DataFrame({'cols':x_train.columns,'box_cox_lambdas':pt.lambdas_})  # checking lambda of all column
```

```python
# Applying Regressing on transformed data
lr = LinearRegression()
lr.fit(x_train_pt,y_train)
y_pred = lr.predict(x_test_pt)
accuracy = r2_score(y_test,y_pred)
accuracy
```

Converting Transformed Data to data frame to plot graphs


```python
x_train_df = pd.DataFrame(x_train_pt,columns=x_train.columns)
x_test_df = pd.DataFrame(x_test_pt,columns=x_test.columns)
```

# Applying yeo-johnson default method

```python
# Applying yeo-johnson default method
pt = PowerTransformer()
x_train_pt = pt.fit_transform(x_train)
x_test_pt = pt.transform(x_test)

lr = LinearRegression()
lr.fit(x_train_pt,y_train)
y_pred = lr.predict(x_test_pt)
accuracy = r2_score(y_test,y_pred)
accuracy
```
